```python
import csv
from rich import print
from scholarly import scholarly
```

```python
author = scholarly.search_author_id('-TInzSAAAAAJ')
# scholarly.pprint(author)

# Retrieve the author's data, fill-in.
author = scholarly.fill(author)

papers = author['publications']
print(papers)

print("Author's papers saved to author_papers.csv")
```

```python
from markdown_table_generator import generate_markdown, table_from_string_list, Alignment

rows = [['Title', 'Citations', 'Year']]
for paper in papers:
    if '*' not in paper['bib']['title']:
        continue

    row = [
        paper['bib']['title'] + ', ' + paper['bib']['citation'],
        paper['num_citations'],
        paper['bib']['pub_year'],
    ]

    row = [str(c) for c in row]
    print(row)

    rows.append(row)

table1 = table_from_string_list(rows, Alignment.LEFT)
markdown1 = generate_markdown(table1)
print(markdown1)
```

```python
from jinja2 import Environment, FileSystemLoader

env = Environment(loader=FileSystemLoader('.'))
template = env.from_string("Hello {{ name }}!")
# %%
template.render(name="John Doe")
# %%
template = env.get_template('index_template.md')

# %%
with open('index.md', 'w', encoding='utf-8') as f:
    f.write(template.render(
        table1=markdown1,
        table2=markdown2,
        today=formatted_date,
    ))
```
